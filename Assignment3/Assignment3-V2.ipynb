Question 1) Please read the "sales.csv" file as a DataFrame object named as "sales_df". Print the length and the first 5 tuples of the DataFrame. (5 points)

```python
# Reading the sales.csv file 
sales_df_org = pd.read_csv("sales.csv")
sales_df = sales_df_org.copy()
print("Length of the sales_df: {}".format(len(sales_df)))
sales_df.head()
```

Question2) Split the dataset into the independent (name it 'x') and dependent variables (name it 'y') which are predictors and response variable( 'Item_Outlet_Sales'). Print their shape info and top 10 records of them. (5 points)

```python
# Determining the x and y variables
x = sales_df.drop(["Item_Outlet_Sales"], axis=1)
y = sales_df[["Item_Outlet_Sales"]]

# Printing the shape of them
print("Shape of independent variable x is: {}".format(x.shape))
print("Shape of dependent variable y is: {}".format(y.shape))
```

```python
x.head(10) # First 10 elements of x
```

```python
y = pd.DataFrame(y)
y.head(10) # First 10 elements of y
```

Question3) Normalization step: Rescale independent variables (x variable) into the range [0,1] by applying  MinMaxScaler to normalize. Lastly convert into dataframe (10 points)

```python
# Importing MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Rescaling the x variable into the range [0,1]
min_max_scaler = MinMaxScaler(feature_range=(0,1))
x_normalized_df = pd.DataFrame(min_max_scaler.fit_transform(x))
x_normalized_df.head()
```

Question 4) Split data into train and test sets. Use the split ratio as 80:20 and set the 'random_state' to an integer value. (10 points)

```python
# Importing Train test split
from sklearn.model_selection import train_test_split
```

```python
# Spliting the data into train and test sets with ratio 80:20
X_train, X_test, y_train, y_test = train_test_split(x_normalized_df, y, test_size=0.2, random_state=1)
print("Shape of X_train: {}".format(X_train.shape))
print("Shape of X_test: {}".format(X_test.shape))
print("Shape of y_train: {}".format(y_train.shape))
print("Shape of y_test: {}".format(y_test.shape))
```

Question 5) Create an instance of kNN with a number of neighbors=3, then fit kNN regression model to train data. Finally, make prediction and calculate mean squared error (MSE). Print MSE value. (15 points)

```python
#importing KNN regressor and metric mse
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as mse
```

```python
knn_regressor = KNN(n_neighbors=3).fit(X_train, y_train) # Create a kNN regression model with number of neighbors = 3
y_pred = knn_regressor.predict(X_test) # Making predictions
mserror = mse(y_test, y_pred) # Calculating and printing the mse
print("MSE value for this model is: {}".format(mserror))
```

Question 6) Create Elbow function which does the same operation in the previous question for each k instance in the range [1-40] by using Manhattan and Euclidean distance metrics separately in order to determine the optimal value of k. (25 points)

```python
# Initialize the k values, 2 lists in order to store the error values
sample_k_values = list(range(1,41))
manhattan_distances, euclidean_distances = [], []
y_test = np.array(y_test) # DataFrame to np.array

# elbow function takes 3 arguments: distribution-type, test values and predictions
def elbow(dist_type, test, predict):

    # This inner function uses the manhattan metric for kNN model
    def manhattan(test, predict):
        for num in sample_k_values:
            knn_regressor = KNN(n_neighbors=num, metric="manhattan").fit(X_train, y_train)
            y_pred = knn_regressor.predict(X_test)
            mserror = mse(y_test, y_pred)
            manhattan_distances.append(mserror) # Storing the error values to the list

    # This inner function uses the euclidean metric for kNN model
    def euclidean(test, predict):
        for num in sample_k_values:
            knn_regressor = KNN(n_neighbors=num, metric="euclidean").fit(X_train, y_train)
            y_pred = knn_regressor.predict(X_test)
            mserror = mse(y_test, y_pred)
            euclidean_distances.append(mserror) # Storing the error values to the list

    # Determining the metric type by the input argument
    if dist_type == "manhattan":
        return manhattan(test, predict)
    elif dist_type == "euclidean":
        return euclidean(test, predict)
```

Call the defined function for defined range of k

```python
# This function call part may take some time
elbow("manhattan", y_test, y_pred)
elbow("euclidean", y_test, y_pred)

# Accessing the indexes of minimum value of error for each list seperately 
# Since the index values begins from zero, we have to increment the value by 1 
optimal_val_manh = manhattan_distances.index(min(manhattan_distances)) + 1
optimal_val_eucl = euclidean_distances.index(min(euclidean_distances)) + 1

#Printing the optimal k values seperately
print("Optimal value of k with Manhattan Distance: {}".format(optimal_val_manh))
print("Optimal value of k with Euclidean Distance: {}".format(optimal_val_eucl))
```

Question 7) Visualize the test mean squared error (MSE) results of Manhattan and Euclidean distance metric by plotting. Two curves should be displayed in the same graph. (15 points)

```python
# Plotting the graphs of mse results for each distance metric in one graph
plt.figure(figsize=(15,8))
plt.plot(sample_k_values, manhattan_distances, label = "Manhattan Distance")
plt.plot(sample_k_values, euclidean_distances, label = "Euclidean Distance")
plt.xlabel("Sample K Values With Range [1,40]", fontsize=15)
plt.ylabel("Mean Squared Error Values", fontsize=15)
plt.title("Mean Squared Error Values For Each KNN Model", fontsize=20)
plt.legend()
plt.show()
```

Question 8) According to obtained result, decide distance metric and optimal k value, and then use them for your kNN model, and make a prediction. Print MSE value for chosen k value. (15 points)

```python
# As we can see the graph, Euclidean distance metric has lower mse value than the Manhattan distance metric.
# So we will use Euclidean distance metric in out kNN model with 9 neighbors that we decided before at question-6
```
