Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
# Firstly I created an empty list to store the values.
happiness_degree_list = []

# By looking at the conditions above, appending the values to the list.
for ladder in range(len(happiness_df.index)):
    ladder_value = happiness_df.iloc[ladder]["Ladder"]

    if (ladder_value < 30):
        happiness_degree_list.append("Very Happy")

    elif (ladder_value >= 30 and ladder_value < 60):
        happiness_degree_list.append("Happy")

    elif (ladder_value >= 60 and ladder_value < 90):
        happiness_degree_list.append("Moderately Happy")

    elif (ladder_value >= 90 and ladder_value < 120):
        happiness_degree_list.append("Unhappy")

    elif (ladder_value >= 120):
        happiness_degree_list.append("Very Unhappy")

        # Finally this list is taking part of the new dataframe.
happiness_df["happiness_degree"] = happiness_degree_list
happiness_df.head(10)
```

Question 9) Plot multiple bar charts in a single chart that show the mean values of 'Social support','Freedom','Generosity' according to happiness degree (15 points)

```python
# Firstly I created empty lists to store the data
social_support, freedom, generosity = [], [], []
social_support_mean, freedom_mean, generosity_mean = [], [], []

# In order to prevent the duplicates, I did this type conversion and sorted the values
happiness_degree_set = sorted(list(set(happiness_degree_list)))
happiness_degree_set[0], happiness_degree_set[3] = happiness_degree_set[3], happiness_degree_set[0]
happiness_degree_set[2], happiness_degree_set[1] = happiness_degree_set[1], happiness_degree_set[2]
happiness_degree_set[3], happiness_degree_set[1] = happiness_degree_set[1], happiness_degree_set[3]

# Collecting the data which is wanted from me from the dataframe
for degree in happiness_degree_set:
    for index, row in happiness_df.iterrows():
        if row["happiness_degree"] == degree:
            social_support.append(row["Social support"])
            freedom.append(row["Freedom"])
            generosity.append(row["Generosity"])

    # Then calculating the mean of these variables and storing them
    social_support_mean.append(np.mean(social_support))
    freedom_mean.append(np.mean(freedom))
    generosity_mean.append(np.mean(generosity))
    social_support.clear()
    freedom.clear()
    generosity.clear()

# In order to plot multiple barplots, creating the new dataframe with these datas which we obtained
multiple_barplot_df = pd.DataFrame({"Social support":social_support_mean, 
                                    "Freedom":freedom_mean, 
                                    "Generosity":generosity_mean}, index=happiness_degree_set)

multiple_barplot_df.plot(kind="bar",figsize=(15, 8))
```

Question 10) Draw random samples 30%, 40%, 50%, 60% of the data with replacement and investigate how the correlation of sample to the whole data changes with respect to sample size. (15 points)

```python
# Obtaining random samples by using the given rates
sample1 = happiness_df.sample(frac=.3, random_state=1, replace=True)
sample2 = happiness_df.sample(frac=.4, random_state=1, replace=True)
sample3 = happiness_df.sample(frac=.5, random_state=1, replace=True)
sample4 = happiness_df.sample(frac=.6, random_state=1, replace=True)

# Checking if the rates are true or not
print(len(happiness_df))
print(len(sample1))
print(len(sample2))
print(len(sample3))
print(len(sample4))
```

```python
# Correlation table for each samples
corr_sample1 = sample1.corr()
corr_sample2 = sample2.corr()
corr_sample3 = sample3.corr()
corr_sample4 = sample4.corr()

# Function that takes 2 arguments which are the sample and the sample's number in order to plotting the heatmap
def heatmapper(sample, sample_num):
    plt.subplots(figsize=(12,8))
    plt.title("Correlations for Sample {}".format(sample_num), fontsize = 20)
    sns.heatmap(sample, cmap="YlOrRd", annot=True, fmt=".3f")
    plt.show()

# And for better understanding, plotting heatmaps for each of them
heatmapper(corr_sample1, 1)
heatmapper(corr_sample2, 2)
heatmapper(corr_sample3, 3)
heatmapper(corr_sample4, 4)
```

Question 11) Plot pie charts for random sampled data obtained in previous question according to happiness_degree. (10 points)

```python
# Color selecting
colors = sns.color_palette()[0:5]

# This function takes 2 arguments which are the sample and the sample's number in order to plot the pie chart
# By using the sample datas, obtaining how many number of each value in happiness_degree column(value count)
# Saving the information as a dictionary
# Keys of the dictionary are the labels of the pie chart
# Values of the dictionary are the values(percentages) of the pie chart
# And recalling the function for each sample
def pie_plotter(sample, sample_num):
    data_sample = dict(sample["happiness_degree"].value_counts())
    plt.subplots(figsize=(12,8))
    plt.title("Pie Chart for Sample {}".format(sample_num), fontsize = 20)
    plt.pie(data_sample.values(), labels = data_sample.keys(), colors = colors, autopct='%.2f%%')
    plt.show()

pie_plotter(sample1, 1)
pie_plotter(sample2, 2)
pie_plotter(sample3, 3)
pie_plotter(sample4, 4)
```
